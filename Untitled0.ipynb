```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import io # <-- Import the 'io' module

# =============================================================================
# Part 1: Setup and Data Loading
# =============================================================================

print("--- Part 1: Loading and Preparing Data ---")

# The raw CSV data from the URL is now embedded here as a string.
csv_data = """Post ID,User ID,Post Content,Platform,Post Type,Likes,Comments,Shares,Date,Time,Followers at Posting,Campaign Name
1,1,"Feeling blessed to see another beautiful sunrise. #blessed #sunrise","Facebook","Photo",150,30,15,"2024-01-01","07:00:00",1000,"New Year Kickstart"
2,1,"Just finished a 5k run! What a great way to start the year. #fitness #running","Instagram","Video",250,50,25,"2024-01-02","08:00:00",1010,"Health & Wellness"
3,2,"My new blog post is live! Check it out: [link] #tech #blogging","LinkedIn","Text",80,15,10,"2024-01-02","09:00:00",500,"Tech Tuesdays"
4,3,"Loving the new features on this app! #innovation #tech","Twitter","Text",50,5,2,"2024-01-03","10:00:00",2000,"Product Launch"
5,2,"Happy to announce our new partnership! Great things ahead. #business #partnership","LinkedIn","Text",300,40,20,"2024-01-04","11:00:00",520,"Corporate Announcements"
6,4,"Our Q4 report is out and the results are amazing! #growth #success","LinkedIn","Text",400,60,30,"2024-01-05","12:00:00",8000,"Quarterly Report"
7,1,"Throwback to our amazing vacation last summer. #tbt #travel","Instagram","Photo",500,80,40,"2024-01-06","13:00:00",1050,"Travel Goals"
8,5,"Just launched my new course on data science. Enroll now! #datascience #education","LinkedIn","Video",600,100,50,"2024-01-07","14:00:00",15000,"Course Launch"
9,6,"This new AI tool is a game-changer for content creators. #AI #creativity","Twitter","Text",120,20,8,"2024-01-08","15:00:00",3000,"AI Advancements"
10,7,"What are your favorite productivity hacks? Share them below! #productivity #tips","Facebook","Text",90,25,5,"2024-01-09","16:00:00",500,"Community Engagement"
11,8,"Behind the scenes of our latest photoshoot. #bts #fashion","Instagram","Photo",700,120,60,"2024-01-10","17:00:00",25000,"Fashion Week"
12,9,"Excited to be speaking at the upcoming conference! #publicspeaking #conference","LinkedIn","Text",250,35,15,"2024-01-11","18:00:00",7000,"Event Promotion"
13,10,"A perfect cup of coffee to start the day. #coffee #morning","Instagram","Photo",350,60,30,"2024-01-12","07:30:00",1200,"Daily Moments"
14,1,"Check out this amazing recipe I tried today! #foodie #recipe","Facebook","Video",450,70,35,"2024-01-13","08:30:00",1060,"Food Fridays"
15,2,"Reflecting on a productive week. #worklife #motivation","LinkedIn","Text",150,25,12,"2024-01-14","09:30:00",530,"Weekly Roundup"
16,3,"The future of remote work is here to stay. #remotework #futureofwork","Twitter","Text",80,10,4,"2024-01-15","10:30:00",2050,"Industry Trends"
17,4,"Our new update is live! Let us know what you think. #update #feedback","Facebook","Text",200,45,18,"2024-01-16","11:30:00",8100,"Product Launch"
18,5,"Honored to be featured in this month's magazine. #achievement #datascience","LinkedIn","Photo",800,150,75,"2024-01-17","12:30:00",15500,"Milestone Celebration"
19,6,"AI ethics is a topic we should all be discussing. #AI #ethics","Twitter","Text",180,30,12,"2024-01-18","13:30:00",3100,"Thought Leadership"
20,7,"Planning my next travel adventure. Where should I go? #travel #adventure","Instagram","Photo",600,100,50,"2024-01-19","14:30:00",550,"Community Engagement"
21,8,"A day in the life of a data scientist. #dayinthelife #datascience","Instagram","Video",900,180,90,"2024-01-20","15:30:00",25500,"Behind the Scenes"
22,9,"Networking is key to career growth. #networking #career","LinkedIn","Text",350,50,25,"2024-01-21","16:30:00",7200,"Career Advice"
23,10,"Sunday vibes. #sunday #relax","Facebook","Photo",250,40,20,"2024-01-22","17:30:00",1250,"Daily Moments"
24,1,"Starting a new week with a positive mindset. #mondaymotivation #positivity","Facebook","Text",180,35,14,"2024-01-23","08:00:00",1070,"New Year Kickstart"
25,2,"Just published a new article on data visualization. #dataviz #analytics","LinkedIn","Text",450,65,32,"2024-01-24","09:00:00",550,"Tech Tuesdays"
26,3,"What's your favorite coding language and why? #coding #programming","Twitter","Text",100,15,6,"2024-01-25","10:00:00",2100,"Community Engagement"
27,4,"We're hiring! Check out our open positions. #hiring #jobs","LinkedIn","Text",500,80,40,"2024-01-26","11:00:00",8200,"Recruitment Drive"
28,5,"My top 5 books for data scientists. #books #datascience","LinkedIn","Photo",700,120,60,"2024-01-27","12:00:00",15800,"Content Marketing"
29,6,"The impact of AI on healthcare is profound. #AI #healthcare","Twitter","Text",220,40,18,"2024-01-28","13:00:00",3200,"AI Advancements"
30,7,"Dreaming of this beach right now. #beach #vacation","Instagram","Photo",800,130,65,"2024-01-29","14:00:00",600,"Travel Goals"
31,8,"Hosting a live Q&A session this Friday. Join me! #live #qa","Instagram","Video",1000,200,100,"2024-01-30","15:00:00",26000,"Community Engagement"
32,9,"The importance of mentorship in career development. #mentorship #career","LinkedIn","Text",400,60,30,"2024-01-31","16:00:00",7300,"Career Advice"
33,10,"Cozy evening with a good book. #reading #cozy","Facebook","Photo",300,50,25,"2024-02-01","17:00:00",1300,"Daily Moments"
34,1,"New month, new goals. #february #goals","Facebook","Text",220,40,16,"2024-02-02","08:00:00",1080,"New Year Kickstart"
35,2,"Exploring the power of machine learning in finance. #ml #finance","LinkedIn","Text",600,90,45,"2024-02-03","09:00:00",560,"Tech Tuesdays"
36,3,"The latest trends in web development for 2024. #webdev #trends","Twitter","Text",150,25,10,"2024-02-04","10:00:00",2150,"Industry Trends"
37,4,"Our team is growing! Welcome to our new members. #team #growth","LinkedIn","Photo",300,50,25,"2024-02-05","11:00:00",8300,"Corporate Announcements"
38,5,"A deep dive into neural networks. #neuralnetworks #AI","LinkedIn","Video",900,180,90,"2024-02-06","12:00:00",16000,"Course Launch"
39,6,"The ethical implications of AI are more important than ever. #AI #ethics","Twitter","Text",250,50,20,"2024-02-07","13:00:00",3300,"Thought Leadership"
40,7,"Winter wonderland. #winter #snow","Instagram","Photo",900,150,75,"2024-02-08","14:00:00",650,"Seasonal"
41,8,"How to build a compelling data story. #datastorytelling #dataviz","Instagram","Video",1200,250,125,"2024-02-09","15:00:00",26500,"Content Marketing"
42,9,"Success is a journey, not a destination. #success #motivation","LinkedIn","Text",500,70,35,"2024-02-10","16:00:00",7400,"Career Advice"
43,10,"Sunday brunch goals. #brunch #foodie","Facebook","Photo",400,60,30,"2024-02-11","17:00:00",1350,"Daily Moments"
44,1,"Happy Valentine's Day! #valentinesday #love","Facebook","Photo",500,80,40,"2024-02-12","08:00:00",1100,"Seasonal"
45,2,"The role of data in business strategy. #data #business","LinkedIn","Text",700,100,50,"2024-02-13","09:00:00",570,"Tech Tuesdays"
46,3,"The debate: Tabs vs. Spaces. What's your choice? #coding #debate","Twitter","Text",200,80,15,"2024-02-14","10:00:00",2200,"Community Engagement"
47,4,"Excited to announce our new feature that will revolutionize your workflow. #newfeature #innovation","Facebook","Video",600,120,60,"2024-02-15","11:00:00",8400,"Product Launch"
48,5,"The art of clean code. #cleancode #programming","LinkedIn","Text",800,130,65,"2024-02-16","12:00:00",16200,"Content Marketing"
49,6,"How AI is transforming the customer service industry. #AI #customerservice","Twitter","Text",300,60,25,"2024-02-17","13:00:00",3400,"AI Advancements"
50,7,"Take me back to the mountains. #mountains #hiking","Instagram","Photo",1000,180,90,"2024-02-18","14:00:00",700,"Travel Goals"
51,8,"My journey into data science: The early days. #tbt #datascience","Instagram","Photo",1500,300,150,"2024-02-19","15:00:00",27000,"Behind the Scenes"
52,9,"The power of a positive company culture. #companyculture #workplace","LinkedIn","Text",600,90,45,"2024-02-20","16:00:00",7500,"Corporate Announcements"
53,10,"A quiet moment of reflection. #mindfulness #peace","Facebook","Photo",350,55,28,"2024-02-21","17:00:00",1400,"Daily Moments"
54,1,"Feeling grateful for this community. #community #gratitude","Facebook","Text",280,45,22,"2024-02-22","08:00:00",1120,"Community Engagement"
55,2,"Big data challenges and solutions. #bigdata #analytics","LinkedIn","Text",900,140,70,"2024-02-23","09:00:00",580,"Tech Tuesdays"
56,3,"What are the must-have tools for developers in 2024? #devtools #programming","Twitter","Text",250,50,20,"2024-02-24","10:00:00",2250,"Community Engagement"
57,4,"Our commitment to sustainability. #sustainability #gogreen","LinkedIn","Photo",400,70,35,"2024-02-25","11:00:00",8500,"Corporate Social Responsibility"
58,5,"Building machine learning models from scratch. #ml #python","LinkedIn","Video",1100,220,110,"2024-02-26","12:00:00",16500,"Course Launch"
59,6,"The future of autonomous vehicles. #AI #automotive","Twitter","Text",400,80,30,"2024-02-27","13:00:00",3500,"AI Advancements"
60,7,"Chasing sunsets. #sunset #travel","Instagram","Photo",1200,200,100,"2024-02-28","14:00:00",750,"Travel Goals"
61,8,"An inside look at our data infrastructure. #dataengineering #bigdata","Instagram","Video",1800,350,175,"2024-02-29","15:00:00",27500,"Behind the Scenes"
62,9,"How to negotiate a higher salary. #salary #careeradvice","LinkedIn","Text",800,120,60,"2024-03-01","16:00:00",7600,"Career Advice"
63,10,"Weekend vibes. #weekend #fun","Facebook","Photo",500,80,40,"2024-03-02","17:00:00",1450,"Daily Moments"
64,1,"Hello March! New beginnings. #march #newmonth","Facebook","Text",300,50,25,"2024-03-03","08:00:00",1150,"Seasonal"
65,2,"The ethics of AI in decision making. #AI #ethics","LinkedIn","Text",1000,160,80,"2024-03-04","09:00:00",590,"Thought Leadership"
66,3,"The rise of serverless architecture. #serverless #cloud","Twitter","Text",300,60,25,"2024-03-05","10:00:00",2300,"Industry Trends"
67,4,"Celebrating International Women's Day. #IWD2024 #womenintech","LinkedIn","Photo",700,100,50,"2024-03-06","11:00:00",8600,"Corporate Social Responsibility"
68,5,"From data to decisions: A case study. #casestudy #dataanalytics","LinkedIn","Text",1200,200,100,"2024-03-07","12:00:00",16800,"Content Marketing"
69,6,"AI-powered drug discovery is revolutionizing medicine. #AI #medicine","Twitter","Text",500,100,40,"2024-03-08","13:00:00",3600,"AI Advancements"
70,7,"Lost in the beauty of nature. #nature #outdoors","Instagram","Photo",1400,250,125,"2024-03-09","14:00:00",800,"Travel Goals"
71,8,"Our new interactive dashboard is live! #dataviz #dashboard","Instagram","Video",2000,400,200,"2024-03-10","15:00:00",28000,"Product Launch"
72,9,"How to build a personal brand. #personalbranding #career","LinkedIn","Text",900,140,70,"2024-03-11","16:00:00",7700,"Career Advice"
73,10,"A simple pleasure: a walk in the park. #simplepleasures #park","Facebook","Photo",450,70,35,"2024-03-12","17:00:00",1500,"Daily Moments"
74,1,"Spring is in the air! #spring #flowers","Facebook","Photo",600,90,45,"2024-03-13","08:00:00",1180,"Seasonal"
75,2,"The future of quantum computing. #quantum #technology","LinkedIn","Text",1300,220,110,"2024-03-14","09:00:00",600,"Tech Tuesdays"
76,3,"What's the most underrated programming language? #programming #discussion","Twitter","Text",350,70,28,"2024-03-15","10:00:00",2350,"Community Engagement"
77,4,"Our annual company retreat was a huge success! #teamwork #retreat","LinkedIn","Video",800,120,60,"2024-03-16","11:00:00",8700,"Corporate Announcements"
78,5,"A practical guide to A/B testing. #abtesting #marketing","LinkedIn","Text",1400,250,125,"2024-03-17","12:00:00",17000,"Content Marketing"
79,6,"AI is helping us fight climate change. #AI #climatechange","Twitter","Text",600,120,50,"2024-03-18","13:00:00",3700,"AI Advancements"
80,7,"Urban exploration. #cityscape #urban","Instagram","Photo",1600,280,140,"2024-03-19","14:00:00",850,"Travel Goals"
81,8,"Behind the algorithm: How our recommendation system works. #algorithm #recommendations","Instagram","Video",2500,500,250,"2024-03-20","15:00:00",28500,"Behind the Scenes"
82,9,"The art of giving and receiving feedback. #feedback #communication","LinkedIn","Text",1000,160,80,"2024-03-21","16:00:00",7800,"Career Advice"
83,10,"Nothing beats a homemade meal. #homecooking #food","Facebook","Photo",550,85,42,"2024-03-22","17:00:00",1550,"Daily Moments"
84,1,"Looking forward to the weekend. #weekend #almostthere","Facebook","Text",350,60,30,"2024-03-23","08:00:00",1200,"Community Engagement"
85,2,"The intersection of AI and creativity. #AI #creativity","LinkedIn","Text",1500,280,140,"2024-03-24","09:00:00",610,"Thought Leadership"
86,3,"The best VS Code extensions for productivity. #vscode #devtools","Twitter","Text",400,80,35,"2024-03-25","10:00:00",2400,"Community Engagement"
87,4,"We're proud to support our local community. #community #support","LinkedIn","Photo",900,140,70,"2024-03-26","11:00:00",8800,"Corporate Social Responsibility"
88,5,"Deploying machine learning models at scale. #mlops #deployment","LinkedIn","Video",1800,350,175,"2024-03-27","12:00:00",17500,"Course Launch"
89,6,"The role of AI in personalized education. #AI #education","Twitter","Text",700,140,60,"2024-03-28","13:00:00",3800,"AI Advancements"
90,7,"A walk to remember. #memories #walk","Instagram","Photo",1800,320,160,"2024-03-29","14:00:00",900,"Travel Goals"
91,8,"Our top 10 data science projects of the year. #datascience #projects","Instagram","Video",3000,600,300,"2024-03-30","15:00:00",29000,"Content Marketing"
92,9,"How to overcome impostor syndrome. #impostorsyndrome #mentalhealth","LinkedIn","Text",1200,200,100,"2024-03-31","16:00:00",7900,"Career Advice"
93,10,"The beauty of a quiet morning. #morning #peace","Facebook","Photo",650,100,50,"2024-04-01","17:00:00",1600,"Daily Moments"
94,1,"Happy April Fools' Day! #aprilfools #joke","Facebook","Text",400,150,80,"2024-04-02","08:00:00",1220,"Seasonal"
95,2,"Natural Language Processing (NLP) explained simply. #NLP #AI","LinkedIn","Text",1800,320,160,"2024-04-03","09:00:00",620,"Tech Tuesdays"
96,3,"Is low-code the future of development? #lowcode #nocode","Twitter","Text",500,100,45,"2024-04-04","10:00:00",2450,"Industry Trends"
97,4,"Our commitment to diversity and inclusion. #diversity #inclusion","LinkedIn","Photo",1000,160,80,"2024-04-05","11:00:00",8900,"Corporate Social Responsibility"
98,5,"Advanced SQL techniques for data analysts. #SQL #dataanalysis","LinkedIn","Text",2000,400,200,"2024-04-06","12:00:00",18000,"Course Launch"
99,6,"How AI is helping to create more sustainable cities. #AI #smartcities","Twitter","Text",800,160,70,"2024-04-07","13:00:00",3900,"AI Advancements"
100,7,"Exploring the hidden gems of my city. #city #explore","Instagram","Photo",2000,350,175,"2024-04-08","14:00:00",950,"Travel Goals"
"""

# Use io.StringIO to let pandas read the string as a file.
df = pd.read_csv(io.StringIO(csv_data))

# Renaming columns for easier use in SQL and Python.
df.columns = [
    'post_id', 'user_id', 'post_content', 'platform', 'post_type', 'likes',
    'comments', 'shares', 'date', 'time', 'followers_at_posting', 'campaign_name'
]

# Converting 'date' column to datetime objects for time series analysis.
df['date'] = pd.to_datetime(df['date'])

# Creating a 'total_engagement' helper column.
df['total_engagement'] = df['likes'] + df['comments'] + df['shares']

print("Data loaded and prepared successfully!")
print("\nFirst 5 rows of the dataset:")
print(df.head())

# =============================================================================
# Part 2: SQL Implementation
# =============================================================================

print("\n\n--- Part 2: SQL Analysis ---")

# SQL Plan: Create table and use GROUP BY for aggregation.
conn = sqlite3.connect(':memory:')
df.to_sql('posts', conn, index=False, if_exists='replace')

# Task: Use SQL to calculate average likes, shares, comments by platform and campaign.
query_platform_avg = """
SELECT
    platform,
    AVG(likes) AS avg_likes,
    AVG(comments) AS avg_comments,
    AVG(shares) AS avg_shares
FROM
    posts
GROUP BY
    platform
ORDER BY
    avg_likes DESC;
"""
df_platform_avg = pd.read_sql_query(query_platform_avg, conn)
print("\nAverage Engagement by Platform (from SQL):")
print(df_platform_avg)

query_campaign_avg = """
SELECT
    campaign_name,
    AVG(likes) AS avg_likes,
    AVG(comments) AS avg_comments,
    AVG(shares) AS avg_shares,
    SUM(total_engagement) as total_engagement
FROM
    posts
GROUP BY
    campaign_name
ORDER BY
    total_engagement DESC;
"""
df_campaign_avg = pd.read_sql_query(query_campaign_avg, conn)
print("\nAverage Engagement by Campaign (from SQL):")
print(df_campaign_avg)

# --- Calculating KPIs from the project description ---
query_kpis = """
SELECT
    /* Engagement Rate: (Likes + Shares + Comments) / Total Followers */
    SUM(total_engagement) * 1.0 / SUM(followers_at_posting) AS engagement_rate,
    /* Average Engagement per Post: AVG(Likes + Shares + Comments) */
    AVG(total_engagement) AS avg_engagement_per_post,
    /* Share Rate: Shares / Total Posts (equivalent to AVG(Shares)) */
    AVG(shares) AS share_rate,
    /* Comment Rate: Comments / Total Posts (equivalent to AVG(Comments)) */
    AVG(comments) AS comment_rate
FROM
    posts;
"""
df_kpis = pd.read_sql_query(query_kpis, conn)
print("\nOverall Key Performance Indicators (KPIs):")
print(df_kpis)

# Close the database connection
conn.close()

# =============================================================================
# Part 3: Python Implementation & Visualization
# =============================================================================

print("\n\n--- Part 3: Python Analysis & Visualization ---")

# --- 1. Visualize Engagement by Platform (Stacked Bar Chart) ---
platform_engagement = df.groupby('platform')[['likes', 'comments', 'shares']].sum()
platform_engagement.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='viridis'
)
plt.title('Total Engagement by Social Media Platform', fontsize=16)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Total Count (Likes + Comments + Shares)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Engagement Type')
plt.tight_layout()
plt.show()

# --- 2. Visualize Engagement Trends Over Time (Line Plot) ---
engagement_over_time = df.groupby('date')['total_engagement'].sum().reset_index()
plt.figure(figsize=(14, 7))
sns.lineplot(data=engagement_over_time, x='date', y='total_engagement', marker='o')
plt.title('Daily Social Media Engagement Trend', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Engagement', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 3. Visualize Engagement by Post Type (Bar Plot) ---
plt.figure(figsize=(12, 7))
sns.barplot(
    data=df,
    x='post_type',
    y='total_engagement',
    hue='platform',
    estimator='mean',
    ci=None,
    palette='magma'
)
plt.title('Average Engagement by Post Type and Platform', fontsize=16)
plt.xlabel('Post Type', fontsize=12)
plt.ylabel('Average Engagement per Post', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Platform')
plt.tight_layout()
plt.show()
```
